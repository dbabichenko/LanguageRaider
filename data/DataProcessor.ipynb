```python
f = open("conversation.xml")
lines = f.readlines()
lines
```

```python
phrases = []
for line in lines:
    if "<phrase is" in line:
        phrases.append(line.strip())

phrases
```

```python
scenario_id = 1
cnt = 1
for phrase in phrases:
    temp = phrase.split('"')
    sql = "INSERT INTO statements (scenarioID, statementID, statementText, isCorrect) VALUES ("
    sql += "'" + str(scenario_id) + "', "
    sql += "'" + str(cnt) + "', "
    sql += "'" + temp[3] + "', "
    if temp[1] == 'true':
        sql += "1 "
    else:
        sql += "0 "
    sql = sql + ");"
    cnt = cnt + 1
    #print(temp[1] + " : " + temp[3])
    print(sql)
```
